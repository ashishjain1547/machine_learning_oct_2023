```python
import pandas as pd
import numpy as np
import math
from mpmath import gamma
```

```python
df = pd.read_excel('est earnings 29 Sep 2020 - 29 Sep 2023/report.xlsx')
```

```python
df['inflated_earning'] = df['Estimated earnings (USD)'] * 10
```

```python
df
```

```python
lm = (1 - math.pi) * df['inflated_earning'].mean()
print("lambda: " + str(lm))
k_list = np.linspace(1, 15, num=15)


def get_poisson_probability(lm, k):
    ret = (1 - math.pi) * math.pow(lm, k) * math.pow(math.e, -lm) / float(gamma(k))
    return ret

kl = []
pl = []

kl.append(0)
pl.append(math.pi + (1 - math.pi) * math.pow(math.e, -lm))

for i in k_list:
    kl.append(round(i / 10, 2))
    pl.append(round(get_poisson_probability(lm, i), 4))

df_report = pd.DataFrame({
    'k': kl,
    'probability': pl
})

df_report
```

```python
plt.scatter(df_report['k'], df_report['probability'])
plt.show()
```

```python
lm = df['Estimated earnings (USD)'].mean()
print("lambda: " + str(lm))
k_list = np.linspace(0.05, 1.5, num=30)


def get_poisson_probability(lm, k):
    ret = math.pow(lm, k) * math.pow(math.e, -lm) / float(gamma(k))
    return ret

kl = []
pl = []

kl.append(0)
pl.append(math.pow(lm, 0) * math.pow(math.e, -lm) / float(math.factorial(0)))

for i in k_list:
    kl.append(round(i, 2))
    pl.append(get_poisson_probability(lm, i))

df_report = pd.DataFrame({
    'k': kl,
    'probability': pl
})

df_report
```

```python
plt.scatter(df_report['k'], df_report['probability'])
plt.show()
```
