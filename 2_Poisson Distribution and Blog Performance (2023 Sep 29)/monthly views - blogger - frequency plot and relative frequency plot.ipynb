```python
import pandas as pd
import mpmath
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
bins = pd.cut(df['Views'], 10) # Equal frequency binning
```

```python
df['Bins'] = bins
```

```python
df
```

```python
df_rep = df.groupby(by = ['Bins'], as_index = False)['Views'].count()
```

```python
df_rep['Bins'] = df_rep['Bins'].astype(str)
```

```python
import matplotlib.pyplot as plt

plt.title("Frequency Plot for Views")
plt.xlabel("Page Views Buckets")
plt.xticks(rotation=90)
plt.bar(df_rep['Bins'], df_rep['Views'])
plt.show() 
```
